Lambda Function

```python
add = lambda x , y : x + y
print(add(8, 8))
```

```python
square = lambda x : x**2
print(square(4))
```

```python
Even_Odd = lambda x: "Even" if x % 2 == 0 else "Odd"
print(Even_Odd(3))
```

```python
multiple = lambda x , y, z: x * y + z
print(multiple(22,32,43))
```

Map

```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
num = [2,4,3,21,4,54]
result = list(map(lambda x : x + 1, num))
print(result)
```

filter

```python
l1 = [45, 65, 90, 33, 55]
result = list(filter(lambda x: x % 2 == 0 , l1))
print(result)
```
